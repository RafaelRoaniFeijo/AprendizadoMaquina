O treinamento (incluindo a seleção de modelos no spot-checking e a otimização de hiperparâmetros) acontece em porções (folds) do seu DATASET_FILE (sintético+real). A cada iteração do loop outer_cv_train_splitter, 80% desse DATASET_FILE é usado para treinar um modelo.
A avaliação final de cada um desses modelos treinados é feita usando 100% do seu REAL_DATASET_FILE como um conjunto de teste/validação fixo e independente.
Portanto, a porcentagem de 80/20 se aplica à divisão interna do DATASET_FILE para criar diferentes conjuntos de treinamento através da validação cruzada, mas a avaliação final é sempre contra o REAL_DATASET_FILE completo.

```python
# =====================================
# 1. IMPORTAÇÕES E CONFIGURAÇÃO GERAL
# =====================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import time
from scipy.stats import t, friedmanchisquare, wilcoxon, ttest_rel
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, label_binarize
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
import xgboost as xgb
from sklearn.metrics import f1_score, confusion_matrix, classification_report, roc_curve, auc, precision_score, recall_score, brier_score_loss
from sklearn.model_selection import StratifiedKFold, RepeatedStratifiedKFold, RandomizedSearchCV, cross_val_score, learning_curve
from sklearn.pipeline import Pipeline
from sklearn.inspection import permutation_importance
from scipy.stats import randint, uniform

from sklearn.calibration import calibration_curve
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

warnings.filterwarnings('ignore')

# Para rodar no Colab, monte o drive
# from google.colab import drive
# drive.mount('/content/drive')
# DATASET_FILE = '/content/drive/My Drive/Dataset_OriginalSinteticosComClass_PREPROCESSADO_COMPLETO.csv'
# REAL_DATASET_FILE = '/content/drive/My Drive/Dataset_OriginalComClass_PREPROCESSADO_COMPLETO.csv'


# Local:
# Dataset para TREINO (Sintético + Real)
DATASET_FILE = "G:\Meu Drive\.mestrado\Aprendizado de máquina\AtividadePraticaKNN\PreProcessados\Dataset_OriginalSinteticosComClass_PREPROCESSADO_COMPLETO.csv"
# Dataset para VALIDAÇÃO/TESTE (Real)
REAL_DATASET_FILE = "G:\Meu Drive\.mestrado\Aprendizado de máquina\AtividadePraticaKNN\PreProcessados\Dataset_OriginalComClass_PREPROCESSADO_COMPLETO.csv"


TARGET      = "Adequação MILHO"
RANDOM_SEED = 42
TOP_N       = 5
N_SPLITS_OUTER = 5 # Define em quantos folds do dataset "sintético+real" o treino será feito
N_SPLITS_INNER = 5
N_REPEATS_HPO  = 5

le      = LabelEncoder() # LabelEncoder será ajustado nos dados de treino
imputer = SimpleImputer(strategy="median")

scalers = {
    "NoExplicitScaler": None,
    "STD": StandardScaler
}

base_models = {
    "k-NN":               KNeighborsClassifier(),
    "DecisionTree":       DecisionTreeClassifier(class_weight="balanced", random_state=RANDOM_SEED),
    "RandomForest":       RandomForestClassifier(class_weight="balanced", random_state=RANDOM_SEED),
    "SVM":            SVC(kernel="rbf", class_weight="balanced", probability=True, random_state=RANDOM_SEED),
    "XGBoost":            xgb.XGBClassifier(use_label_encoder=False, eval_metric="mlogloss", random_state=RANDOM_SEED),
}

param_dists = {
    "k-NN":               {"model__n_neighbors":randint(3,50), "model__p":[1,2], "model__weights":["uniform","distance"]},
    "DecisionTree":       {"model__max_depth":randint(3,50), "model__min_samples_split":randint(2,10), "model__min_samples_leaf":randint(1,5), "model__criterion":["gini","entropy"]},
    "RandomForest":       {"model__n_estimators":randint(50,300), "model__max_depth":randint(5,50), "model__min_samples_split":randint(2,10), "model__min_samples_leaf":randint(1,5), "model__criterion":["gini","entropy"]},
    "SVM":            {"model__C": uniform(0.1,10), "model__gamma":["scale","auto"] + list(np.logspace(-3,2,6))},
    "XGBoost":            {"model__n_estimators":randint(50,300), "model__max_depth":randint(3,10), "model__learning_rate":uniform(0.01,0.2),
                           "model__subsample":uniform(0.6,0.4), "model__colsample_bytree":uniform(0.6,0.4)}
}

# Carregar dataset de TREINO (Sintético + Real)
try:
    df_train_full = pd.read_csv(DATASET_FILE, sep=';', decimal=',')
    print(f"Dataset de TREINO '{DATASET_FILE}' carregado com sucesso. Shape: {df_train_full.shape}")
except FileNotFoundError:
    print(f"ERRO: Arquivo de TREINO '{DATASET_FILE}' não encontrado. Por favor, defina o caminho correto.")
    exit()

X_train_full = df_train_full.drop(columns=[TARGET])
y_train_full_labels = df_train_full[TARGET]
# Ajustar o LabelEncoder APENAS nos dados de treino (sintético+real)
y_train_full_encoded = le.fit_transform(y_train_full_labels)
class_names = le.classes_.astype(str) # Nomes das classes baseados no dataset de treino
n_unique_classes = len(class_names)

# Carregar dataset de VALIDAÇÃO/TESTE (Real)
try:
    df_test_real = pd.read_csv(REAL_DATASET_FILE, sep=';', decimal=',')
    print(f"Dataset de VALIDAÇÃO/TESTE '{REAL_DATASET_FILE}' carregado com sucesso. Shape: {df_test_real.shape}")
except FileNotFoundError:
    print(f"ERRO: Arquivo de VALIDAÇÃO/TESTE '{REAL_DATASET_FILE}' não encontrado. Por favor, defina o caminho correto.")
    exit()

X_test_real = df_test_real.drop(columns=[TARGET])
y_test_real_labels = df_test_real[TARGET]
# Transformar os labels do dataset REAL usando o LabelEncoder JÁ AJUSTADO
try:
    y_test_real_encoded = le.transform(y_test_real_labels)
except ValueError as e:
    print(f"ERRO ao transformar labels do dataset de VALIDAÇÃO/TESTE '{REAL_DATASET_FILE}': {e}")
    print("Provavelmente há classes no dataset de VALIDAÇÃO/TESTE que não estavam no dataset de TREINO.")
    print(f"Classes aprendidas pelo LabelEncoder (do treino): {list(le.classes_)}")
    print(f"Classes únicas no dataset de VALIDAÇÃO/TESTE: {list(y_test_real_labels.unique())}")
    print("Considere ajustar o LabelEncoder em todas as classes possíveis ou filtrar o dataset de VALIDAÇÃO/TESTE.")
    exit()


all_results = {sc_name: {model_name: {'fold_metrics': [], 'confusion_matrices': [], 'execution_times_hpo': [], 'best_estimators': []}
                                for model_name in base_models.keys()}
                     for sc_name in scalers.keys()}
best_roc_data_storage = {sc_name: {model_name: {'best_f1': -1, 'y_test_actual': None,
                                                'y_pred_probabilities': None, 'best_fold_index': -1,
                                                'n_classes_for_roc': n_unique_classes,
                                                'class_names_for_roc': class_names} # class_names do treino
                                      for model_name in base_models.keys()}
                           for sc_name in scalers.keys()}

# ==================================================
# 2. PIPELINE PRINCIPAL DE TREINO E AVALIAÇÃO
# ==================================================

for sc_name, ScalerCls in scalers.items():
    current_scaler_results = all_results[sc_name]

    if ScalerCls is not None:
        pipelines_spot_checking = {
            name: Pipeline([('imputer', imputer), ('scaler', ScalerCls()), ('model', mdl)])
            for name, mdl in base_models.items()
        }
        print_header_main_loop = f"===== ESCALONAMENTO: {sc_name} ====="
        print_spot_checking_model_template = "Spot-checking {model_name} (escalador: {scaler_name}, CV em dados de treino)"
        print_selected_hpo_models_template = "Modelos selecionados para HPO com escalonador {scaler_name}"
    else:
        pipelines_spot_checking = {
            name: Pipeline([('imputer', imputer), ('model', mdl)])
            for name, mdl in base_models.items()
        }
        print_header_main_loop = f"===== CONFIGURAÇÃO SEM ESCALONADOR EXPLÍCITO: {sc_name} ====="
        print_spot_checking_model_template = "Spot-checking {model_name} (config: {scaler_name}, CV em dados de treino)"
        print_selected_hpo_models_template = "Modelos selecionados para HPO (configuração {scaler_name})"

    print(f"\n\n{print_header_main_loop}\n")

    # Spot-checking continua a ser feito por CV no dataset de TREINO COMPLETO (sintético+real)
    spot_scores_mean = {}
    cv_spot_check = StratifiedKFold(n_splits=N_SPLITS_INNER, shuffle=True, random_state=RANDOM_SEED)
    print("--- Iniciando Spot-Checking (CV no dataset de TREINO 'sintético+real') ---")
    for name, pipe_sc in pipelines_spot_checking.items():
        # Usa X_train_full e y_train_full_encoded para spot-checking
        scores = cross_val_score(pipe_sc, X_train_full, y_train_full_encoded, cv=cv_spot_check, scoring='f1_macro', n_jobs=-1)
        spot_scores_mean[name] = scores.mean()
        print_spot_msg = print_spot_checking_model_template.format(model_name=name, scaler_name=sc_name)
        print(f"{print_spot_msg}: F1_macro médio = {scores.mean():.3f}")
    print("--- Fim do Spot-Checking ---")

    selected_for_hpo = [n for n, _ in sorted(spot_scores_mean.items(), key=lambda x: x[1], reverse=True)[:TOP_N]]
    print_hpo_select_msg = print_selected_hpo_models_template.format(scaler_name=sc_name)
    print(f"\n→ {print_hpo_select_msg}: {selected_for_hpo}")

    # outer_cv define os folds de TREINO a partir do dataset "sintético+real"
    # O dataset de TESTE será sempre o "real" completo
    outer_cv_train_splitter = StratifiedKFold(n_splits=N_SPLITS_OUTER, shuffle=True, random_state=RANDOM_SEED)

    print(f"\n--- Iniciando Treino e Avaliação (Treino em 'sintético+real', Validação em 'real') ---")
    for fold_idx, (train_fold_indices, _) in enumerate(outer_cv_train_splitter.split(X_train_full, y_train_full_encoded)):
        print(f"\n--- Fold de Treino Externo {fold_idx + 1}/{N_SPLITS_OUTER} (Config: {sc_name}) ---")

        # Dados de treino para este fold (do dataset "sintético+real")
        X_train_fold_data = X_train_full.iloc[train_fold_indices]
        y_train_fold_data = y_train_full_encoded[train_fold_indices]

        # Dados de VALIDAÇÃO/TESTE são sempre o dataset "real" completo
        X_test_eval_data = X_test_real.copy() # Usar cópia para segurança
        y_test_eval_data = y_test_real_encoded.copy()

        best_estimators_this_outer_fold = {}
        for model_name in selected_for_hpo:
            print(f"  Processando modelo: {model_name}")
            if ScalerCls is not None:
                current_pipeline = Pipeline([('imputer', imputer), ('scaler', ScalerCls()), ('model', base_models[model_name])])
            else:
                current_pipeline = Pipeline([('imputer', imputer), ('model', base_models[model_name])])

            current_params = param_dists.get(model_name)

            if not current_params: # Modelos sem HPO
                start_time_fit = time.time()
                current_pipeline.fit(X_train_fold_data, y_train_fold_data)
                end_time_fit = time.time()
                best_estimators_this_outer_fold[model_name] = current_pipeline
                current_scaler_results[model_name]['best_estimators'].append(current_pipeline)
                # Armazenar tempo de fit simples se não houver HPO
                current_scaler_results[model_name]['execution_times_hpo'].append(end_time_fit - start_time_fit)
                print(f"    Modelo {model_name} treinado (sem HPO) em {(end_time_fit - start_time_fit):.2f}s.")
                trained_model_for_eval = current_pipeline
            else: # Modelos com HPO
                cv_hpo = RepeatedStratifiedKFold(n_splits=N_SPLITS_INNER, n_repeats=N_REPEATS_HPO, random_state=RANDOM_SEED)
                rs = RandomizedSearchCV(current_pipeline, current_params, scoring='f1_macro',
                                        cv=cv_hpo, n_iter=30, n_jobs=-1,
                                        random_state=RANDOM_SEED, refit=True)
                print(f"    Iniciando HPO para {model_name}...")
                start_time_hpo = time.time()
                rs.fit(X_train_fold_data, y_train_fold_data)
                end_time_hpo = time.time()
                print(f"    HPO para {model_name} concluído em {(end_time_hpo - start_time_hpo):.2f}s. Melhor F1 (interno CV): {rs.best_score_:.4f}")

                best_estimators_this_outer_fold[model_name] = rs.best_estimator_
                current_scaler_results[model_name]['best_estimators'].append(rs.best_estimator_)
                current_scaler_results[model_name]['execution_times_hpo'].append(end_time_hpo - start_time_hpo)
                trained_model_for_eval = rs.best_estimator_

            # Avaliação no dataset "real" (X_test_eval_data, y_test_eval_data)
            print(f"    Avaliando {model_name} no dataset REAL...")
            y_pred_on_real = trained_model_for_eval.predict(X_test_eval_data)

            f1_macro_real = f1_score(y_test_eval_data, y_pred_on_real, average='macro', zero_division=0)
            precision_macro_real = precision_score(y_test_eval_data, y_pred_on_real, average='macro', zero_division=0)
            recall_macro_real = recall_score(y_test_eval_data, y_pred_on_real, average='macro', zero_division=0)
            # labels=range(n_unique_classes) usa as classes do TREINO. Se REAL tiver menos classes, alguns labels podem não aparecer.
            # É importante que y_test_eval_data use os mesmos encodings numéricos.
            cm_real = confusion_matrix(y_test_eval_data, y_pred_on_real, labels=le.transform(le.classes_))
            report_dict_real = classification_report(y_test_eval_data, y_pred_on_real, target_names=class_names, output_dict=True, zero_division=0, labels=le.transform(le.classes_))
            print(f"    F1 Macro (REAL) para {model_name}: {f1_macro_real:.4f}")

            metrics_dict = {
                'fold_index': fold_idx, # Este é o índice do fold de TREINO
                'f1_macro': f1_macro_real,
                'precision_macro': precision_macro_real,
                'recall_macro': recall_macro_real,
                'report_dict': report_dict_real,
                'confusion_matrix': cm_real
            }
            current_scaler_results[model_name]['fold_metrics'].append(metrics_dict)
            current_scaler_results[model_name]['confusion_matrices'].append(cm_real)

            if hasattr(trained_model_for_eval.named_steps['model'], "predict_proba"):
                y_pred_probas_fold_real = trained_model_for_eval.predict_proba(X_test_eval_data)
                # Armazenar dados ROC se este fold deu o melhor F1 no dataset REAL
                if f1_macro_real > best_roc_data_storage[sc_name][model_name]['best_f1']:
                    best_roc_data_storage[sc_name][model_name]['best_f1'] = f1_macro_real
                    best_roc_data_storage[sc_name][model_name]['y_test_actual'] = y_test_eval_data # Labels do dataset REAL
                    best_roc_data_storage[sc_name][model_name]['y_pred_probabilities'] = y_pred_probas_fold_real
                    best_roc_data_storage[sc_name][model_name]['best_fold_index'] = fold_idx
                    # n_classes_for_roc e class_names_for_roc já estão definidos com base no TREINO, o que é correto para consistência.

# ==========================================
# 3. FUNÇÕES AUXILIARES PARA ANÁLISES
# (Funções existentes como boxplot_metric, compare_models_stat_test, etc., permanecem as mesmas)
# Elas operarão sobre as métricas coletadas, que agora são da validação no dataset REAL.
# ==========================================

def boxplot_metric(models_results, metric='f1_macro', config_name='Default'):
    data = []
    labels = []
    for model_name, results in models_results.items():
        vals = [m[metric] for m in results['fold_metrics']]
        if vals:
            data.append(vals)
            labels.append(model_name)
    if not data:
        print(f"Sem dados para plotar boxplot para métrica {metric} na configuração {config_name}.")
        return
    plt.figure(figsize=(10, 6))
    plt.boxplot(data, labels=labels, patch_artist=True)
    plt.title(f'Boxplot {metric} por Modelo ({config_name})\n(Validação no Dataset REAL)')
    plt.ylabel(f"{metric} (no Dataset REAL)")
    plt.xticks(rotation=45, ha="right")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def compare_models_stat_test(models_results, metric='f1_macro', config_name='Default'):
    model_scores = {model_name: [m[metric] for m in results['fold_metrics']]
                    for model_name, results in models_results.items() if results['fold_metrics']}

    valid_models = {k: v for k, v in model_scores.items() if len(v) > 1}

    if len(valid_models) < 2:
        print(f"Não há modelos suficientes com resultados para comparação estatística ({metric}) na config {config_name} (Val. REAL).")
        return

    print(f"\nComparação Estatística ({metric}) para Config: {config_name} (Val. REAL)")

    f1_arrays = list(valid_models.values())
    same_length = all(len(arr) == len(f1_arrays[0]) for arr in f1_arrays)

    if len(f1_arrays) > 2 and same_length:
        try:
            stat, p = friedmanchisquare(*f1_arrays)
            print(f"Teste de Friedman p-valor: {p:.4f}")
            if p < 0.05:
                print("  Há diferença estatisticamente significativa entre os modelos.")
            else:
                print("  Não há diferença estatisticamente significativa entre os modelos segundo Friedman.")
        except ValueError as e:
            print(f"Não foi possível realizar o teste de Friedman: {e}")
    elif len(f1_arrays) == 2 and same_length:
         model_names_list = list(valid_models.keys())
         print(f"Comparando {model_names_list[0]} e {model_names_list[1]}:")
         try:
            t_stat, t_p = ttest_rel(f1_arrays[0], f1_arrays[1])
            print(f"  Teste T pareado: p-valor = {t_p:.4f}")
         except ValueError as e:
            print(f"  Não foi possível realizar o Teste T pareado: {e}")
         try:
            w_stat, w_p = wilcoxon(f1_arrays[0], f1_arrays[1], zero_method='zsplit')
            print(f"  Teste de Wilcoxon: p-valor = {w_p:.4f}")
         except ValueError as e:
            print(f"  Não foi possível realizar o Teste de Wilcoxon: {e}")
    else:
        print("Não foi possível realizar o teste de Friedman (requer >2 grupos com o mesmo número de observações por grupo ou tamanhos diferentes).")
        print("Comparações par a par (Wilcoxon) para todos os pares de modelos com mesmo número de observações:")
        model_names_list = list(valid_models.keys())
        for i in range(len(model_names_list)):
            for j in range(i + 1, len(model_names_list)):
                model1_name = model_names_list[i]
                model2_name = model_names_list[j]
                scores1 = valid_models[model1_name]
                scores2 = valid_models[model2_name]
                if len(scores1) == len(scores2) and len(scores1) > 1:
                    try:
                        _, p_wilcoxon = wilcoxon(scores1, scores2, zero_method='zsplit')
                        print(f"  Wilcoxon entre {model1_name} e {model2_name}: p-valor={p_wilcoxon:.4f}")
                    except ValueError as e:
                        print(f"  Não foi possível realizar Wilcoxon entre {model1_name} e {model2_name}: {e}")

# Curvas de aprendizado ainda usam X_train_full, y_train_full_encoded para mostrar aprendizado no domínio de treino.
def learning_curve_plot(estimator, X, y, model_name="Modelo", config_name="Default"):
    try:
        # Nota: X e y aqui devem ser do domínio de treino (sintético+real)
        train_sizes, train_scores, test_scores = learning_curve(
            estimator, X, y, cv=StratifiedKFold(n_splits=N_SPLITS_INNER, shuffle=True, random_state=RANDOM_SEED),
            scoring='f1_macro', train_sizes=np.linspace(0.1, 1.0, 5), n_jobs=-1)

        plt.figure(figsize=(8, 6))
        plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', color="r", label='Score de Treino (em subset de treino)')
        plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', color="g", label='Score de Val. Cruzada (em subset de treino)')

        plt.title(f'Curva de Aprendizagem - {model_name} ({config_name})\n(Baseado no Dataset de TREINO)')
        plt.xlabel('Número de Amostras de Treino')
        plt.ylabel('F1 Macro Score')
        plt.legend(loc="best")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Erro ao gerar curva de aprendizagem para {model_name} ({config_name}): {e}")

# Métricas por classe serão baseadas na validação no dataset REAL.
def class_report_aggregate(models_results, config_name='Default'):
    print(f"\nMétricas Agregadas por Classe para Config: {config_name} (Val. REAL)")
    for model_name, model_data in models_results.items():
        if model_data['fold_metrics']:
            print(f"\nModelo: {model_name}")
            metrics_accumulator = {cls_name: {'precision':[], 'recall':[], 'f1-score':[]} for cls_name in class_names} # class_names do treino

            for fold_data in model_data['fold_metrics']:
                report = fold_data['report_dict']
                for cls_idx, cls_name_from_train in enumerate(class_names): # Iterar sobre classes do treino
                    # Report pode não ter todas as classes se o dataset REAL não as tiver
                    if cls_name_from_train in report:
                        metrics_accumulator[cls_name_from_train]['precision'].append(report[cls_name_from_train]['precision'])
                        metrics_accumulator[cls_name_from_train]['recall'].append(report[cls_name_from_train]['recall'])
                        metrics_accumulator[cls_name_from_train]['f1-score'].append(report[cls_name_from_train]['f1-score'])
                    # else: # Caso a classe do treino não esteja no report do REAL (e.g. 0 amostras)
                        # print(f"    Aviso: Classe '{cls_name_from_train}' (do treino) não encontrada no report do dataset REAL para o fold.")

            for cls_name in class_names:
                print(f"  Classe: {cls_name}")
                for metric_name in ['precision', 'recall', 'f1-score']:
                    vals = metrics_accumulator[cls_name][metric_name]
                    if vals:
                        print(f"    {metric_name.capitalize()}: {np.mean(vals):.3f} ± {np.std(vals):.3f} (de {len(vals)} folds)")
                    else:
                        print(f"    {metric_name.capitalize()}: N/A (sem dados nos reports dos folds para esta classe)")
        else:
            print(f"\nModelo: {model_name} - Sem métricas de fold para agregar (Val. REAL).")

# Curvas de calibração: X, y_true podem ser X_test_real, y_test_real_encoded para ver calibração no dataset REAL.
# O estimador (clf_pipeline) foi treinado no sintético+real.
def calibration_curves_plot(clf_pipeline, X_calib, y_calib_true, model_name="Modelo", config_name="Default"):
    actual_model = clf_pipeline.named_steps['model']
    if not hasattr(actual_model, "predict_proba"):
        print(f"Modelo {model_name} ({config_name}) não suporta predict_proba. Curva de calibração não gerada.")
        return

    print(f"\nCurvas de Calibração e Brier Score - {model_name} ({config_name})\n(Avaliando no dataset fornecido para calibração)")

    try:
        # O pipeline (clf_pipeline) já tem imputer e scaler (se aplicável),
        # então X_calib será processado por ele.
        y_pred_probs = clf_pipeline.predict_proba(X_calib)

        plt.figure(figsize=(10, 8))
        overall_brier = 0

        # Usar class_names do LabelEncoder global (ajustado no treino)
        for i, class_label in enumerate(class_names):
            # y_calib_true já deve estar encodado numericamente (0, 1, ...)
            y_true_class_binary = (y_calib_true == le.transform([class_label])[0]).astype(int)

            class_probs = y_pred_probs[:, i] # Coluna i corresponde à classe i do treino

            fraction_of_positives, mean_predicted_value = calibration_curve(y_true_class_binary, class_probs, n_bins=10, strategy='uniform')

            plt.plot(mean_predicted_value, fraction_of_positives, "s-", label=f"Classe {class_label}")

            brier = brier_score_loss(y_true_class_binary, class_probs, pos_label=1)
            print(f"  Brier Score para Classe {class_label}: {brier:.4f}")
            overall_brier += brier

        plt.plot([0, 1], [0, 1], "k--", label="Calibração Perfeita")
        plt.xlabel("Probabilidade Média Prevista")
        plt.ylabel("Fração de Positivos")
        plt.legend(loc="lower right")
        plt.title(f"Curvas de Calibração - {model_name} ({config_name})")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

        if len(class_names) > 0:
             print(f"  Brier Score Médio Geral: {(overall_brier / len(class_names)):.4f}")
        else:
             print("  Não foi possível calcular Brier Score Médio Geral (sem classes).")


    except Exception as e:
        print(f"Erro ao gerar curvas de calibração para {model_name} ({config_name}): {e}")
        import traceback
        traceback.print_exc()


def tradeoff_plot(models_results, config_name='Default'):
    # Esta função usa 'execution_times_hpo' e métricas dos folds (que são da validação no REAL)
    models = []
    mean_f1 = []
    mean_time_hpo = []

    for model_name, model_data in models_results.items():
        if model_data['fold_metrics'] and model_data['execution_times_hpo']:
            models.append(model_name)
            mean_f1.append(np.mean([m['f1_macro'] for m in model_data['fold_metrics']]))
            mean_time_hpo.append(np.mean(model_data['execution_times_hpo']))
        elif model_data['fold_metrics'] and not model_data['execution_times_hpo']:
             print(f"Aviso: {model_name} ({config_name}) tem métricas (Val. REAL) mas não tem tempos HPO para trade-off.")

    if not models:
        print(f"Sem dados para plotar gráfico de trade-off para config {config_name} (Val. REAL).")
        return

    plt.figure(figsize=(10, 6))
    safe_max_time = max(mean_time_hpo) if mean_time_hpo else 0
    safe_max_f1 = max(mean_f1) if mean_f1 else 0

    plt.scatter(mean_time_hpo, mean_f1, s=100, alpha=0.7, edgecolors="k")

    for i, model_label in enumerate(models):
        plt.text(mean_time_hpo[i] + (safe_max_time * 0.01 if safe_max_time > 0 else 0.01),
                 mean_f1[i] + (safe_max_f1 * 0.01 if safe_max_f1 > 0 else 0.01),
                 model_label, fontsize=9)

    plt.xlabel('Tempo Médio de Execução HPO (segundos)')
    plt.ylabel('F1 Macro Médio (no Dataset REAL)')
    plt.title(f'Trade-off: F1 Macro vs. Tempo de HPO ({config_name})\n(Validação no Dataset REAL)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# SMOTE e Ensemble são análises no dataset de TREINO (sintético+real)
def smote_scores(X_train_data, y_train_data_encoded): # Modificado para aceitar dados de treino
    print("\nResultados com SMOTE (aplicado aos dados de TREINO 'sintético+real', CV interna):")
    for name, mdl in base_models.items():
        pipe_smote = ImbPipeline([
            ('imputer', SimpleImputer(strategy="median")),
            ('scaler', StandardScaler()),
            ('smote', SMOTE(random_state=RANDOM_SEED)),
            ('model', mdl)
        ])
        try:
            # CV no dataset de treino fornecido
            scores = cross_val_score(pipe_smote, X_train_data, y_train_data_encoded,
                                     cv=StratifiedKFold(n_splits=N_SPLITS_INNER, shuffle=True, random_state=RANDOM_SEED),
                                     scoring='f1_macro', n_jobs=-1)
            print(f"  {name} + SMOTE: F1 Macro (CV em treino) = {np.mean(scores):.3f} ± {np.std(scores):.3f}")
        except Exception as e:
            print(f"  Erro ao calcular SMOTE score para {name}: {e}")

def ensemble_voting(X_train_data, y_train_data_encoded): # Modificado para aceitar dados de treino
    print("\nResultados com Ensemble VotingClassifier (CV nos dados de TREINO 'sintético+real'):")
    estimators = [
        ('rf', RandomForestClassifier(class_weight="balanced", random_state=RANDOM_SEED)),
        ('xgb', xgb.XGBClassifier(use_label_encoder=False, eval_metric="mlogloss", random_state=RANDOM_SEED)),
        ('svm', SVC(kernel="rbf", class_weight="balanced", probability=True, random_state=RANDOM_SEED))
    ]

    voting_clf_soft = VotingClassifier(estimators=estimators, voting='soft', n_jobs=-1)
    pipeline_soft = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('scaler', StandardScaler()),
        ('model', voting_clf_soft)
    ])

    try:
        # CV no dataset de treino fornecido
        scores_soft = cross_val_score(pipeline_soft, X_train_data, y_train_data_encoded,
                                      cv=StratifiedKFold(n_splits=N_SPLITS_INNER, shuffle=True, random_state=RANDOM_SEED),
                                      scoring='f1_macro', n_jobs=-1)
        print(f"  VotingClassifier (Soft): F1 Macro (CV em treino) = {np.mean(scores_soft):.3f} ± {np.std(scores_soft):.3f}")
    except Exception as e:
        print(f"  Erro ao calcular VotingClassifier (Soft) score: {e}")

# ================================================
# 4. EXECUÇÃO DAS ANÁLISES ADICIONAIS
# ================================================
run_boxplots = True
run_stats_tests = True
run_per_class_metrics = True
run_tradeoff = True
run_smote = True # Será executado nos dados de TREINO (sintético+real)
run_ensemble = True # Será executado nos dados de TREINO (sintético+real)
run_learning_curves = True # Executado nos dados de TREINO (sintético+real)
run_calibration_curves = True # Pode ser executado no dataset REAL
run_confusion_analysis = True # Métricas são da validação no dataset REAL
run_error_analysis = True # Métricas são da validação no dataset REAL
run_scaler_comparison_analysis = True

# Loop de análise para cada configuração de scaler (NoExplicitScaler, STD)
# Todas as métricas em all_results agora são da validação no dataset REAL
for sc_name_final, scaler_results_final in all_results.items():
    print(f"\n\n{'='*40}\n ANÁLISES PARA CONFIGURAÇÃO: {sc_name_final} \n(Métricas baseadas na VALIDAÇÃO NO DATASET REAL)\n{'='*40}")

    if run_boxplots:
        print(f"\n--- Boxplots das Métricas ({sc_name_final}) ---")
        for metric_to_plot in ['f1_macro', 'precision_macro', 'recall_macro']:
            boxplot_metric(scaler_results_final, metric=metric_to_plot, config_name=sc_name_final)

    if run_stats_tests:
        compare_models_stat_test(scaler_results_final, metric='f1_macro', config_name=sc_name_final)

    if run_per_class_metrics:
        class_report_aggregate(scaler_results_final, config_name=sc_name_final)

    if run_tradeoff:
        tradeoff_plot(scaler_results_final, config_name=sc_name_final)

    # SMOTE e Ensemble são chamados uma vez (para evitar redundância), usando o dataset de TREINO completo.
    if sc_name_final == list(scalers.keys())[0]:
        if run_smote:
            smote_scores(X_train_full.copy(), y_train_full_encoded.copy())
        if run_ensemble:
            ensemble_voting(X_train_full.copy(), y_train_full_encoded.copy())

    # Curvas de Aprendizagem: usam o estimador treinado (em fold de sint.+real) e avaliam em CV no dataset de TREINO completo (sint.+real)
    if run_learning_curves:
        print(f"\n--- Curvas de Aprendizagem ({sc_name_final}, baseadas no dataset de TREINO) ---")
        for model_name_lc in scaler_results_final:
            if scaler_results_final[model_name_lc]['best_estimators']:
                representative_estimator = scaler_results_final[model_name_lc]['best_estimators'][0]
                print(f"Gerando Curva de Aprendizagem para: {model_name_lc}")
                learning_curve_plot(representative_estimator, X_train_full.copy(), y_train_full_encoded.copy(), model_name=model_name_lc, config_name=sc_name_final)
            else:
                print(f"Sem estimadores salvos para {model_name_lc} ({sc_name_final}) para gerar curva de aprendizagem.")

    # Curvas de Calibração: usam o estimador treinado (em fold de sint.+real) e avaliam no dataset REAL completo
    if run_calibration_curves:
        print(f"\n--- Curvas de Calibração ({sc_name_final}, avaliadas no dataset REAL) ---")
        for model_name_calib in scaler_results_final:
            if scaler_results_final[model_name_calib]['best_estimators']:
                representative_estimator_calib = scaler_results_final[model_name_calib]['best_estimators'][0]
                print(f"Gerando Curva de Calibração para: {model_name_calib}")
                calibration_curves_plot(representative_estimator_calib, X_test_real.copy(), y_test_real_encoded.copy(), model_name=model_name_calib, config_name=sc_name_final)
            else:
                print(f"Sem estimadores salvos para {model_name_calib} ({sc_name_final}) para gerar curva de calibração.")

    # Matrizes de confusão e Análise de Erros são baseadas nos resultados da validação no dataset REAL
    if run_confusion_analysis:
        print(f"\n--- Matrizes de Confusão Agregadas ({sc_name_final}, Val. REAL) ---")
        for model_name, model_data in scaler_results_final.items():
            cms = model_data['confusion_matrices']
            if cms:
                valid_cms = [cm for cm in cms if isinstance(cm, np.ndarray) and cm.shape == (n_unique_classes, n_unique_classes)] # Checa forma
                if not valid_cms:
                    print(f"  {model_name}: Sem matrizes de confusão válidas ou com forma esperada ({n_unique_classes}x{n_unique_classes}).")
                    continue
                try:
                    aggregated_cm = np.sum(np.array(valid_cms), axis=0)
                    plt.figure(figsize=(7, 6))
                    sns.heatmap(aggregated_cm, annot=True, fmt="d", cmap="Blues",
                                xticklabels=class_names, yticklabels=class_names, # class_names do treino
                                annot_kws={"size": 10})
                    plt.title(f"Matriz de Confusão Agregada (Val. REAL)\n{model_name} ({sc_name_final})", fontsize=12)
                    plt.ylabel("Classe Real", fontsize=10)
                    plt.xlabel("Classe Predita", fontsize=10)
                    plt.xticks(rotation=45, ha="right")
                    plt.yticks(rotation=0)
                    plt.tight_layout()
                    plt.show()
                except Exception as e:
                    print(f"  Erro ao agregar/plotar matriz de confusão para {model_name} ({sc_name_final}): {e}")
            else:
                print(f"  {model_name} ({sc_name_final}): Sem matrizes de confusão para agregar (Val. REAL).")

    if run_error_analysis:
        print(f"\n--- Análise de Erros ({sc_name_final}, Val. REAL) ---")
        for model_name, model_data in scaler_results_final.items():
            if model_data['fold_metrics']:
                valid_fold_metrics = [
                    fm for fm in model_data['fold_metrics']
                    if 'f1_macro' in fm and 'confusion_matrix' in fm and
                       isinstance(fm['confusion_matrix'], np.ndarray) and
                       fm['confusion_matrix'].shape == (n_unique_classes, n_unique_classes)
                ]
                if not valid_fold_metrics:
                    print(f"\n  {model_name}: Dados de fold insuficientes/inválidos para análise de erros (Val. REAL).")
                    continue
                worst_fold_data = min(valid_fold_metrics, key=lambda x: x['f1_macro'])
                conf_matrix_worst_fold = worst_fold_data['confusion_matrix']

                print(f"\n  Análise de erros para {model_name} (Pior Fold de Treino {worst_fold_data['fold_index']}, F1 Val. REAL: {worst_fold_data['f1_macro']:.3f}):")
                print("  Matriz de confusão (Val. REAL) do pior fold de treino:")
                df_cm_worst = pd.DataFrame(conf_matrix_worst_fold, index=class_names, columns=class_names)
                print(df_cm_worst)

                off_diagonal_cm = conf_matrix_worst_fold.copy()
                np.fill_diagonal(off_diagonal_cm, 0)

                if np.sum(off_diagonal_cm) > 0:
                    most_confused_indices = np.unravel_index(np.argmax(off_diagonal_cm), off_diagonal_cm.shape)
                    real_class_idx, predicted_class_idx = most_confused_indices
                    count = off_diagonal_cm[real_class_idx, predicted_class_idx]

                    if real_class_idx < len(class_names) and predicted_class_idx < len(class_names):
                        print(f"    Principal confusão: Classe Real '{class_names[real_class_idx]}' -> Predita como '{class_names[predicted_class_idx]}' ({count} ocorrências).")
                    else:
                        print(f"    Índices de classe ({real_class_idx}, {predicted_class_idx}) fora do intervalo para class_names.")
                else:
                    print("    Sem erros de classificação (fora da diagonal) nesta matriz.")
            else:
                print(f"\n  {model_name} ({sc_name_final}): Sem métricas de fold para análise de erros (Val. REAL).")
    print(f"\n{'-'*40}\nFIM DAS ANÁLISES PARA CONFIGURAÇÃO: {sc_name_final}\n{'-'*40}\n")

# =======================================================
# 5. ANÁLISE: COMPARAÇÃO STD vs. SEM ESCALONADOR (baseado na Val. REAL)
# =======================================================
def compare_scalers_performance(all_results_data, scaler1_name="STD", scaler2_name="NoExplicitScaler", metric='f1_macro'):
    print(f"\n\n{'='*40}\n COMPARAÇÃO DE PERFORMANCE (Val. REAL): {scaler1_name} vs. {scaler2_name} \n{'='*40}")

    if scaler1_name not in all_results_data or scaler2_name not in all_results_data:
        print(f"ERRO: Resultados para '{scaler1_name}' e/ou '{scaler2_name}' não encontrados. Não é possível comparar.")
        return

    results_scaler1 = all_results_data[scaler1_name]
    results_scaler2 = all_results_data[scaler2_name]

    common_models = set(results_scaler1.keys()) & set(results_scaler2.keys())
    if not common_models:
        print("Nenhum modelo comum encontrado entre as duas configurações para comparar.")
        return

    for model_name in common_models:
        print(f"\n--- Modelo: {model_name} ---")

        scores_s1_list = [m[metric] for m in results_scaler1[model_name].get('fold_metrics', []) if metric in m]
        scores_s2_list = [m[metric] for m in results_scaler2[model_name].get('fold_metrics', []) if metric in m]

        if not scores_s1_list or not scores_s2_list:
            print(f"  Dados de fold ({metric}) insuficientes para {model_name} em '{scaler1_name}' ou '{scaler2_name}'.")
            continue

        mean_s1 = np.mean(scores_s1_list)
        std_s1 = np.std(scores_s1_list)
        mean_s2 = np.mean(scores_s2_list)
        std_s2 = np.std(scores_s2_list)

        print(f"  Média da Métrica ({metric}) na Validação REAL:")
        print(f"    {scaler1_name:<20}: {mean_s1:.4f} ± {std_s1:.4f} (de {len(scores_s1_list)} folds de treino)")
        print(f"    {scaler2_name:<20}: {mean_s2:.4f} ± {std_s2:.4f} (de {len(scores_s2_list)} folds de treino)")

        if len(scores_s1_list) == len(scores_s2_list) and len(scores_s1_list) > 1:
            try:
                stat, p_value = wilcoxon(scores_s1_list, scores_s2_list, zero_method='zsplit', alternative='two-sided')
                print(f"  Teste de Wilcoxon ({scaler1_name} vs. {scaler2_name}): Estatística={stat:.2f}, p-valor={p_value:.4f}")
                if p_value < 0.05:
                    print(f"    ➡ Diferença estatisticamente significativa no {metric} para {model_name}.")
                    if mean_s1 > mean_s2:
                        print(f"       Configuração '{scaler1_name}' apresentou performance significativamente melhor.")
                    elif mean_s2 > mean_s1:
                         print(f"       Configuração '{scaler2_name}' apresentou performance significativamente melhor.")
                    else:
                         print(f"       Médias idênticas, mas p-valor < 0.05 (verificar distribuição dos scores).")
                else:
                    print(f"     Não há diferença estatisticamente significativa no {metric} para {model_name}.")
            except ValueError as e:
                print(f"  Não foi possível realizar o teste de Wilcoxon para {model_name}: {e}")
        else:
            print("  Teste estatístico de Wilcoxon não realizado (número de folds/observações diferente ou insuficiente).")

if run_scaler_comparison_analysis:
    compare_scalers_performance(all_results, scaler1_name="STD", scaler2_name="NoExplicitScaler", metric='f1_macro')

print("\n Todas as configurações e análises foram processadas. ")
```
